### O que é o Rotten Tomatoes?

O Rotten Tomatoes é um site agregador de críticas de filmes e séries de TV famoso na internet pelo seu modelo de classificação: os filmes são avaliados em “tomates frescos” para títulos bons, e “tomates podres” para os que não agradaram tanto.

Chamado de “Tomatometer” (algo como “Tomatômetro”), a nota é exibida em formato de porcentagem e calculada com base em um sistema próprio. Saiba como funciona.

### Como funcionam as notas do Rotten Tomatoes?

O Rotten Tomatoes calcula o Tomatometer de um filme ou série de TV com base nas centenas de críticas de mídias especializadas, como sites, portais e jornais, cadastradas em seu banco de dados. A nota é exibido como uma porcentagem de 0 a 100%, sendo exemplificado com os característicos “tomates” que dão nome ao site.

Quando pelo menos 60% das críticas são positivas, o filme recebe o status de “tomate fresco” (ou Fresh, em inglês), e um ícone de tomate vermelho é exibido na página do título com destaque. Se menos de 60% das críticas forem positivas, o filme recebe o status de “tomate podre” (Rotten, em inglês), e recebe um ícone verde de “mancha”.

O site também classifica com um “selo de qualidade” alguns títulos que atingem os seguintes requisitos:

1. Ter a nota igual ou acima de 75%;
2. Pelo menos 5 avaliações de “Top Critics”, que são os críticos considerados mais importantes pelo Rotten Tomatoes;
3. Títulos exibidos em várias salas de cinema devem ter pelo menos 80 avaliações;
4. Para lançamentos limitados (aqueles que passam em poucos cinemas), o mínimo são 40 avaliações;
5. Apenas temporadas de séries de TV com pelo menos 40 avaliações podem receber o selo (e não a série como um todo).

O Rotten Tomatoes esclarece que estes são apenas os requisitos mínimos para um filme ganhar o status de “Certified Fresh“, e que nem todos os filmes que completam esta lista necessariamente recebem o selo.

### 4.2 Separando a Coluna 'Info'

```python
# Aqui podemos ver como está distribuída a coluna 'info'
df['info']
```

```python
# Primeiro passo para organizar esses dados será de dividí-los:
# Mas antes precisamos deixar esse trabalho um pouco mais fácil

s = df['info'].str.replace(r'\n *\n', '\n', regex = True)
print(s)
```

```python
# Testando como ficaria a divisão em lista.
print(s.str.split('\n')[0])
```

```python
# Agora criaremos uma função que irá nos retornar uma lista com todos os titulos de informação presentes no Dataframe.
def CalculaListaInfo(x):
    lista = []
    if isinstance(x, list):
        for elemento in x:
            valor = elemento.split(':')[0].strip()
            lista.append(valor)
    return lista

propriedades = list(set(s.str.split('\n').apply(CalculaListaInfo).sum()))
```

```python
# Lista com todos os tipos de informação existentes na coluna 'info':

propriedades
```

```python
# Próximo passo é criar as colunas com as informações
# Através de uma função definiremos como serão feitas as divisões e os valores atribuidos para aquela linha.
def ConverteListaParaDicionario(x, propriedade):
    d = {}
    if isinstance(x, list):
        for elemento in x:
            item = elemento.split(':')
            chave = item[0].strip()
            valor = item[1].strip()
            d[chave] = valor
    
    if propriedade in d:
        return d[propriedade]
    else:
        return np.nan

# Por fim aplicamos a função para cada linha das novas recém criadas colunas:
for propriedade in propriedades:
    df[propriedade] = s.str.split('\n').apply(ConverteListaParaDicionario, args = (propriedade,))

df
```

```python
# Agora temos todas as propriedades da coluna 'info' bem separadas e definidas:
df.info()
```

```python
# agora podemos apagar a coluna 'info' do dataset:

df = df.drop(columns = 'info')
df.info()
```

### 4.4 Criando novos datasets

```python
# Criar uma nova coluna 'actor_name_adjusted'
df['actor_name_adjusted'] = ''

for j in range(len(df['actor_name'])):
    atores_str = df['actor_name'][j].replace("[", "").replace("]", "").replace("\\\\n", "").replace("'", "").replace('\\n', '').replace('"','')
    atores = atores_str.split(',')

    for i in range(len(atores)):
        atores[i] = atores[i].strip()

    df['actor_name_adjusted'][j] = atores


print(df['actor_name_adjusted'])
```

```python
df['Director_name_adjusted'] = ''

for j in range(len(df['Director'])):
    diretores = str(df['Director'][j])
    diretores = diretores.split(',')

    for i in range(len(diretores)):
        diretores[i] = diretores[i].strip()

    if 'nan' in diretores:
        diretores = [np.nan]  
    else:
        diretores = [str(diretor) for diretor in diretores] 

    df['Director_name_adjusted'][j] = diretores

# Contar os valores na coluna 'Director_name_adjusted', excluindo os NaN
diretores_contagem = df['Director_name_adjusted'].explode().value_counts(dropna=True)
print(diretores_contagem)
```

Heatmap não mostra correlações relevantes o suficiente para inferir colocações sem entrar numa afirmação de viés.

Podemos ver tamném que os atores, diretores e roteiristas renomados tendem a ter maiores avaliações, porém não existe nenhum dado que corrobore essa tese. 

'Audience x tomatometer', não nos mostra nenhuma grande correlação entre eles, porém ambos tendem a convergir.

Bilheteria geralmente no meio e fim do ano tendem a ter melhores desempenhos, em especial com filmes de ação e aventura, porém não existe quase nenhuma relação com performance no audience score e tomatometer score.

Podemos concluir que bilheteria não é tão importante para determinar as notas, muito embora seja um fator importante de desempenho comercial.

Os dados nos revelaram também que a duração padrão entre 90-100 minutos é algo quase mandatório nessa indústria. Mas que também pouco interfere no desempenho das críticas.

Devido a falta de correlação entre os dados, a melhor forma para afirmar se um filme é bem avaliado no Rotten Tomatoes seria se ambos Crítica e Audiencia concordassem em sua análise de recomendação.
